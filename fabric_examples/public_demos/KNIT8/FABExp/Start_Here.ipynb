# KNIT8 Tutorial: FABRIC Experiment Framework (FABExp)

FABRIC Experiment Framework (FABExp) is a Jupyter-Based Graphical User Interface (GUI) implemented using ipywidgets. Users can take advantage of FABExp to manage their experiments in the FABRIC testbed. 

For the KNIT8 tutorial session, users have different options to access the FABExp packages. 

## Options

- Option1: Use the notebook [FABExp package downloader](./FABExp_Downloader.ipynb) to select the FABExp package, e.g, FABExp_fablib_l3_network_fabnet_IPv4.tar.gz, download it and untar it.

- Option2: Open a terminal, run `wget https://raw.githubusercontent.com/fabric-testbed/FABExp/main/demo/FABExp_fablib_l3_network_fabnet_IPv4.tar.gz`. Once the tarball is downloaded, you can run `tar -xvf FABExp_fablib_l3_network_fabnet_IPv4.tar.gz`

- Option3: Clone the entire Github repo by running `git clone https://github.com/fabric-testbed/FABExp.git`. The package is inside the *demo* folder. Run `tar -xvf FABExp_fablib_l3_network_fabnet_IPv4.tar.gz` there.

Then you can go into the *fablib_create_l3network_fabnet_ipv4_manual* folder and run **welcome.ipynb** to start using FABExp.
