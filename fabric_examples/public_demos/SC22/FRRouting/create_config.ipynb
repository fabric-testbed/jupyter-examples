```python
                    
import json

file_name='config/chameleon_four.json'

router_cores='32'
router_ram='128'
router_disk='10'

edge_cores='16'
edge_ram='64'
edge_disk='10'



ffr_config = { 'slice': {}, 
               'networks': [],
               'links': [], 
                'nodes': []}


sites = [ 'DALL', 'UTAH', 'SALT', 'CHAMELEON']
links = [ ('DALL','CHAMELEON'),
         ('DALL','SALT'),
         ('SALT','CHAMELEON'),
         ('UTAH','SALT'),
        ]
nodes_per_site=2


#Add Chameleon network
subnet= f'192.168.1.0/24'
gw_ip= f'192.168.1.1'
ffr_config['networks'].append({'name': f'chameleon', 
                                'site': 'STAR',  
                                'facility': 'CHI@UC',
                                'subnet': subnet,
                                'allocation_pool_start': '192.168.1.10',
                                'allocation_pool_end': '192.168.1.250',
                                'chameleon_gateway_ip': '192.168.1.2',
                                'router' : {   'name': 'chameleon_router', 
                                               'facility': 'FABRIC',
                                               'site': 'MICH',
                                               'network': 'cham', 
                                               'ip': '192.168.1.1', 
                                               'cores': router_cores, 'ram': router_ram, 'disk': router_disk,
                                               'static_routes': [ ] }
                               })   

for n in range( nodes_per_site):
    ffr_config['nodes'].append({  'name': f'chameleon_node{n}', 
                                   'facility': 'CHI@UC',
                                   'compute_reservation_id': '45f60b56-8ecd-4d3e-aa3e-f02e1bebce01',
                                   'network': 'chameleon', 
                                   'ip': f'192.168.1.{100+n}', 
                                   'cores': edge_cores, 'ram': edge_ram, 'disk': edge_disk,
                                   'static_routes': [ ('192.168.0.0/16', '192.168.1.1') ] 
                               })
                                  


for site in sites:
    if site == 'CHAMELEON': continue
    
    subnet= f'192.168.{sites.index(site)+2}.0/24'
    gw_ip= f'192.168.{sites.index(site)+2}.1'
    
    
    
    #Add other networks
    ffr_config['networks'].append({'name': site.lower(), 
                                    'site': site, 
                                    'facility': 'FABRIC',
                                    'subnet': subnet,
                                    'router' : {   'name': f'{ site.lower()}_router', 
                                                   'facility': 'FABRIC',
                                                   'network': site.lower(), 
                                                   'site': site,
                                                   'ip': gw_ip, 
                                                   'cores': router_cores, 'ram': router_ram, 'disk': router_disk,
                                                   'static_routes': [ ] }
                                  })
    for n in range( nodes_per_site):
        ffr_config['nodes'].append({   'name': f'{site.lower()}_node{n+1}', 
                                       'facility': 'FABRIC',
                                       'network': f'{site.lower()}', 
                                       'ip': f'192.168.{sites.index(site)+2}.{n+100}', 
                                        'cores': edge_cores, 'ram': edge_ram, 'disk': edge_disk,
                                       'static_routes': [ ('192.168.0.0/16', gw_ip) ] 
                                   })


        
        
for link in links:                                   
    ffr_config['links'].append({'name': f'link{links.index(link)}', 
                                'subnet': f'192.168.{links.index(link)+100}.0/24',
                                'endpoints': ({'router_name': f'{link[0].lower()}_router', 'ip': f'192.168.{links.index(link)+100}.1' }, 
                                              {'router_name': f'{link[1].lower()}_router', 'ip': f'192.168.{links.index(link)+100}.2' }) 
                               } )
                                   

                                   
                                   



json_object = json.dumps(ffr_config, indent=4)
 
# Writing to sample.json
with open(file_name, "w") as outfile:
    outfile.write(json_object)

#print(f"{list(map(lambda x: x['name'], ffr_config['links']))}")

#print(f"{json.dumps(ffr_config, indent = 4)}")
            
#frr_experiment = FRRouting_Experiment('FRRouting_OSPF_SC22', config=ffr_config)
```
